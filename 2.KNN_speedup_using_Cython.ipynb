Cython은 C data types을 사용하는 파이썬이다. 실제론 파이썬 코드이지만 C file로써 컴파일된다. 이 라이브러리를 호출하는 것이 파이썬 파일을 호출하는 것보다 빠르다. 얼마나 빠를지 한번 확인해보자~

### simple loop

```python
from time import time

from cython_loop import cython_loop # python3 setup.py build_ext --inplace로 cythonized
from python_loop import python_loop

t1 = time()

python_loop()

t2 = time()

cython_loop()

t3 = time()

print("looping with python takes : " + str(t2 - t1))
print("looping with cython takes : " + str(t3 - t2))
```

Cython이 거의 1000배 빠르다!

# Why use a custom distance? 

데이터를 다룰 때, 더 많은 특징을 사용할 수록 더 좋다. 종종, 연속형 변수와 이산형 변수를 동시에 사용해는 상황이 있다. 이러한 경우, 전통적인 distance는 좋은 결과를 가져다 주지 않았다. 이럴 때, 커스텀 distance가 필요한 것이다. 이러한 distance의 선택지는 셀 수도 없이 많지만 가장 예측력이 좋은 것을 섞어서 고르면 된다. 아래는 사잇런에서 사용되는 일반적인 metrics들이다. 

* Euclidean : sqrt(sum((x-y)^2))
* Manhanttan : sum(|x-y|)
* Chebyshev : max(|x-y|)
* Minkowski with parameter p : sum(|x-y|^p)^(1/p)
* W-Minkowski with parameters p,w : sum(w\*|x-y|^p)^(1/p)
* S-Euclidean with parameter V : sqrt(sum((x-y)^2/V))
* Mahalanobis with parameter V : sqrt((x-y)' V^-1(x-y))

하나의 예로, custom metrics은 추천에 유용하다. 어떤 가격의, 어떤 장소에, 어떤 별(star)을 특징으로 가지는 호텔이라고 생각해보자. 이를 KNN으로 해결하기 위해선 서로 다른 수준의 특징을 비교해야하기 때문에 조심해야 한다. 이를 극복하는 한가지는, 아래 distance를 이용해서 normalize하는 것이다.

$$\alpha‖geoloc_1-geoloc_2‖^2 + \beta‖price_1-price_2‖^2 + \gamma‖stars_1-stars_2‖^2$$

# Custom distance syntax 

가장 먼저 custom distance를 선언하자. cython_metric.pyx 파일에 작성할 것이다. 이 함수는 두 개의 arrays를 인자로 받아 실수로 리턴해줄 것이다.

### cython_metric.pyx 

```python
import numpy as np

def mydist(double[:] x, double[:] y):
    cdef int n = x.shape[0]
    cdef double res = 0
    cdef double resbis = 0
    for i in range(n):
        res += (x[i] - y[i])**2
    return res
```

파이썬에서는 단지 아래와 같이 간단하게 표현 될 것이다.

```python
import numpy as np

def mydist(x, y):
    return np.sum((y-x)**2)
```
